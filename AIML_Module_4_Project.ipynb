```python
class WaterPerceptron:
    def __init__(self, lr=0.01, epochs=50):
        self.lr = lr
        self.epochs = epochs
        self.w = None
        self.b = 0
        self.mistakes = []

    def fit(self, X, y):
        self.w = np.zeros(X.shape[1])

        for epoch in range(self.epochs):
            count = 0

            for i in range(len(y)):

                # Linear output
                prediction = np.dot(X[i], self.w) + self.b

                # Misclassification
                if y[i] * prediction <= 0:
                    self.w = self.w + self.lr * y[i] * X[i]
                    self.b = self.b + self.lr * y[i]
                    count += 1

            self.mistakes.append(count)

    def predict(self, X):
        return np.sign(np.dot(X, self.w) + self.b)
```

```python
model_p = WaterPerceptron(lr=0.01, epochs=50)
model_p.fit(X_train, y_train)

pred_p = model_p.predict(X_test)
```

```python
class GDWaterClassifier:
    def __init__(self, lr=0.001, epochs=500):
        self.lr = lr
        self.epochs = epochs
        self.w = None
        self.b = 0
        self.cost_history = []

    def fit(self, X, y):

        self.w = np.zeros(X.shape[1])
        n = X.shape[0]

        for _ in range(self.epochs):

            # Linear output
            z = np.dot(X, self.w) + self.b

            # Gradients
            dw = (1/n) * np.dot(X.T, (z - y))
            db = (1/n) * np.sum(z - y)

            # Update parameters
            self.w = self.w - self.lr * dw
            self.b = self.b - self.lr * db

            # MSE Cost
            cost = (1/(2*n)) * np.sum((z - y)**2)
            self.cost_history.append(cost)

    def predict(self, X):
        z = np.dot(X, self.w) + self.b
        return np.where(z >= 0, 1, -1)
```

```python
model_gd = GDWaterClassifier(lr=0.001, epochs=500)
model_gd.fit(X_train, y_train)

pred_gd = model_gd.predict(X_test)
```

```python
class MarginWaterClassifier:
    def __init__(self, lr=0.001, lambda_param=0.01, epochs=500):
        self.lr = lr
        self.lambda_param = lambda_param
        self.epochs = epochs
        self.w = None
        self.b = 0

    def fit(self, X, y):

        self.w = np.zeros(X.shape[1])

        for _ in range(self.epochs):

            for i, x_i in enumerate(X):

                condition = y[i] * (np.dot(x_i, self.w) + self.b) >= 1

                if condition:

                    self.w -= self.lr * (2 * self.lambda_param * self.w)

                else:

                    self.w -= self.lr * (2 * self.lambda_param * self.w - x_i * y[i])

                    self.b -= self.lr * (-y[i])

    def predict(self, X):
        linear = np.dot(X, self.w) + self.b
        return np.sign(linear)
```

The labels are converted from {0,1} to {-1,+1} because margin-based classifiers such as the Perceptron and Hinge Loss classifier use the expression

y(w
T
x+b)

where y must be either −1 or +1.

Using ±1:

Makes the classification problem symmetric around zero.
Simplifies the mathematical formulation of the loss function.
Makes gradient computation easier.
Correct predictions produce positive values, while incorrect predictions produce negative values.

Therefore, {-1,+1} is preferred over {0,1} for these algorithms.
